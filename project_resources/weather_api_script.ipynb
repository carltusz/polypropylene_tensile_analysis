```python
import pandas as pd
import numpy as np
import datetime as dt
from datetime import date, timedelta
```

```python
# set up weather history url
url = "https://api.weather.gc.ca/collections/climate-daily/items?"
suffix = "&f=json"

climate_id = "&CLIMATE_IDENTIFIER=6115529"
year_range=list(range(2013,date.today().year+1))

params={
    'CLIMATE_IDENTIFIER':6115529,
    'LOCAL_YEAR':2013
}
```

```python
# https://api.weather.gc.ca/collections/climate-daily/items?CLIMATE_IDENTIFIER=6115525^ORCLIMATE_IDENTIFIER=6115529&f=json
import requests

jsons = []

for year in year_range:
    params={
        'CLIMATE_IDENTIFIER':6115529,
        'LOCAL_YEAR':year
    }
    jsons.append(pd.json_normalize(requests.get(url, params=params).json(),record_path=['features']))
    
jsons[0]
    
```

```python
weather_df=pd.concat(jsons).sort_values('properties.LOCAL_DATE')
weather_df
```

```python
# map NaN to "None"
weather_df=weather_df.replace({'None': np.nan, None:np.nan}, regex=True)
weather_df
```

```python
# want to get 10 day normalization
def ten_day_normals(date_df):
    
    mean_temp = np.mean(date_df['mean_temp'])
    max_temp = max(date_df['max_temp'])
    min_temp=min(date_df['min_temp'])
    swing=max_temp-min_temp
    
    results = {
        'mean': mean_temp,
        'max': max_temp,
        'min': min_temp,
        'swing': swing
    }
    
    return results
    
```

```python
wet_df = clean_weather_df.set_index(['date'])
wet_df
```

```python
# add date column
wet_df['date'] = wet_df.index
wet_df
```

```python
# add end date
wet_df['end_date'] = wet_df['date']+timedelta(days=11)
wet_df
```

```python
wet_df['ten_day_mean'] = wet_df.apply(lambda df: ten_day_normals(wet_df.loc[df['date']:df['end_date']])['mean'] , axis = 1)
wet_df['ten_day_max'] = wet_df.apply(lambda df: ten_day_normals(wet_df.loc[df['date']:df['end_date']])['max'] , axis = 1)
wet_df['ten_day_min'] = wet_df.apply(lambda df: ten_day_normals(wet_df.loc[df['date']:df['end_date']])['min'] , axis = 1)
wet_df['ten_day_swing'] = wet_df.apply(lambda df: ten_day_normals(wet_df.loc[df['date']:df['end_date']])['swing'] , axis = 1)
```

```python
wet_df=wet_df[['date','month','ten_day_mean','ten_day_max','ten_day_min','ten_day_swing']]
wet_df
```

```python
wet_df.to_csv('final_weather.csv', index=False)
```
