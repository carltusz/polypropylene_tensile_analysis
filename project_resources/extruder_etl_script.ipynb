```python
import pandas as pd
from datetime import date
```

```python
ext_one_df = pd.read_csv("extruder_one_7050.csv")
ext_two_df = pd.read_csv("extruder_two_7050.csv")
```

```python
# explore extruder run sheets
ext_one_df['Product'].unique()
```

```python
print(ext_two_df.columns)
print(ext_one_df.columns)
```

```python
# apend run sheets

all_ext_df = pd.concat([ext_one_df,ext_two_df])
```

```python
all_ext_df
```

```python
# want to keep only text that contains '7050'
all_ext_df = all_ext_df[all_ext_df['Product'].str.match('7050')== True]
```

```python
# CHECK FOR NAN
all_ext_df.isna().sum()
```

```python
# drop columns with many NAN
all_ext_df=all_ext_df.drop(columns=['DateCodeDate','TopCoolingR','BottomCoolingR'])
all_ext_df
```

```python
all_ext_df.dtypes
```

```python
# convert date stamps to datetime

all_ext_df['DateTime_Stamp']=pd.to_datetime(all_ext_df['DateTime_Stamp'])
```

```python
# drop redundant dates, other irrelevant info
all_ext_df = all_ext_df.drop(columns=['Recipe_DateTime','ShopOrder'])
```

```python
# remove "Extruder " from Machine
all_ext_df['Machine']=all_ext_df['Machine'].map(lambda x: pd.to_numeric(x[-1]))
```

```python
all_ext_df['TotalRunFootage']=all_ext_df['RunFootage']-all_ext_df['LengthOffset']
```

```python
all_ext_df
```

```python
all_ext_df['date']=pd.to_datetime(all_ext_df['DateTime_Stamp']).dt.date
all_ext_df
```

```python
# establish width from part number
all_ext_df['width']=all_ext_df['PartNumber'].str.extract(r'([\d]{,4})\d\d$').astype(int)
```

```python
# all parts are basically identical (7050 plastic) -> remove part number, description
all_ext_df=all_ext_df.drop(columns=['PartNumber','Product'])
```

```python
# save clean data
all_ext_df.to_csv('final_extruder.csv', index=False)
```
